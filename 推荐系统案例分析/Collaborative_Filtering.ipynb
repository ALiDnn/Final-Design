```python
#读入数据
content = []
with open('./wikilens-ratings/out.wikilens-ratings') as fp:
    content = fp.readlines()
    content.remove(content[0])

content
```

```python
#格式化数据集
data = {}
for line in content:
    line = line.strip().replace('\n','').split('\t')
    if not line[0] in data.keys():
        data[line[0]] = {line[1]:line[2]}
    else:
        data[line[0]][line[1]] = line[2]

data
```

```python
#计算用户相似度
from math import *

def Euclid(user1,user2):

    #取出两个用户购买过的商品
    user1_data = data[user1]
    user2_data = data[user2]
    #默认距离
    distance = 0

    #遍历寻找两个用户都购买过的商品
    for key in user1_data.keys():
        if key in user2_data.keys():
            if float(user1_data[key]) == float(user2_data[key]):
                distance += 1
            else:   
                distance += pow(float(user1_data[key])-float(user2_data[key]),2)

    return 1/(1+sqrt(distance))
```

```python
#计算某个用户与其他用户的相似度比对
def top_similar(user):
    
    res = []
    
    for userid in data.keys():
        #排除当前用户
        if not userid == user:
            similar = Euclid(user,userid)
            res.append((userid,similar))
    
    res.sort(key = lambda val:val[1])

    return res
```

```python
print(top_similar('1'))
```

```python
#构建推荐方法
def recommend(user):
    #取相似度最高的用户
    top_user = top_similar(user)[0][0]
    #相似度最高的用户的购买记录
    items = data[top_user]
    #推荐列表
    recommend_list = []
    for item in items.keys():
        if item not in data[user].keys():
            recommend_list.append((item,items[item]))
    
    #多个推荐商品按照评分排序
    recommend_list.sort(key = lambda val:val[1],reverse=True)

    n = int(len(recommend_list)*0.5)
    
    return recommend_list[:n]
```

```python
print(recommend('1'))
```
