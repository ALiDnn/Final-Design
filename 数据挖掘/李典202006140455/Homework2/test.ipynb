```python
import pandas as pd

store_data = pd.read_csv('supermarket.csv') 
store_data.dropna(inplace = True) 
store_data.head()
```

```python
store_data.drop(columns = 'total',inplace=True)
columns_list = list(store_data)
for i in range(len(store_data)):
    for j in range(len(store_data.columns)):
        if store_data.iloc[i,j] == 't':
            store_data.iloc[i,j] = columns_list[j]
```

```python
store_data
```

```python
store_data = store_data.values.tolist()
# for i in range(len(store_data)):
#     store_data[i] = list(filter(lambda x : x != '?', store_data[i]))
```

```python
records = store_data
records
```

```python
from apyori import apriori
"""
def apriori(transactions, **kwargs):
     Arguments:
         transactions：数据集
         min_support：最小支持度
         min_confidence：最小置信度
         min_lift：最小提升度
         max_length：最小项数.
"""
association_rules = apriori(records, min_support=0.15, min_confidence=0.9)  
association_results = list(association_rules)
print(len(association_results))
```

```python
association_results
```

```python
for item in association_results:
    pair = item
    # print(item[0])
    items = [x for x in pair]
    # print(items[1])
    # 输出关联
    print("Rule: " + str(items[0]) + " -> " + str(items[1]))
    # 输出支持度
    print("Support: " + str(item[1]))
    # 输出确信度
    print("Confidence: " + str(item[2][0][2]))
    
    print("=====================================")
```
