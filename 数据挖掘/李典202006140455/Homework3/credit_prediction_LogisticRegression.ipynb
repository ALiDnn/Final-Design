```python
import pandas as pd

df = pd.read_csv('credit-g.csv')
df
```

```python
#查看缺失数据
df.isnull().sum()
```

```python
print(df.info())
```

异常值处理并将所有类型不为int64的列进行独热编码

```python
n = 0
columns_list = []
for i in df.columns:
    if df.dtypes[n] != 'int64' :
        columns_list.append(i)
    else:
        twice_std1 = df.iloc[:,n].mean() + 2 * df.iloc[:,n].std()
        twice_std2 = df.iloc[:,n].mean() - 2 * df.iloc[:,n].std()
        for j in range(len(df)):
            if df.iloc[j,n] > twice_std1:
                df.iloc[j,n] = twice_std1
            elif df.iloc[j,n] < twice_std2:
                df.iloc[j,n] = twice_std2
    n += 1
df=pd.get_dummies(df,columns = columns_list[:-1])

df
```

数值化'class'列

```python
from sklearn import preprocessing

class_LE = preprocessing.LabelEncoder()
df['class'] = class_LE.fit_transform(df['class'].values)
df
```

划分数据集

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df[[i for i in list(df.columns) if i != 'class']],df['class'],test_size=0.2)
```

标准化

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
x_train = std.fit_transform(x_train)
x_test = std.transform(x_test)
```

使用逻辑回归进行预测

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=1.0)
lr.fit(x_train,y_train)
print(lr.coef_)
y_predict = lr.predict(x_test)
```

```python
print("准确率为：",lr.score(x_test,y_test))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predict,labels=[1,0],target_names=["good","bad"]))
```
