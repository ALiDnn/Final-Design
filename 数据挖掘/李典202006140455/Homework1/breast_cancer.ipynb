```python
from sklearn import preprocessing

age_encoder = preprocessing.LabelEncoder()
menopause_encoder = preprocessing.LabelEncoder()
tumor_size_encoder = preprocessing.LabelEncoder()
inv_nodes_encoder = preprocessing.LabelEncoder()
node_caps_encoder = preprocessing.LabelEncoder()
breast_encoder = preprocessing.LabelEncoder()
breast_quad_encoder = preprocessing.LabelEncoder()
irradiat_encoder = preprocessing.LabelEncoder()
Class_encoder = preprocessing.LabelEncoder()

df['age'] = age_encoder.fit_transform(df['age'].values)
df['menopause'] = menopause_encoder.fit_transform(df['menopause'].values)
df['tumor-size'] = tumor_size_encoder.fit_transform(df['tumor-size'].values)
df['inv-nodes'] = inv_nodes_encoder.fit_transform(df['inv-nodes'].values)
df['node-caps'] = node_caps_encoder.fit_transform(df['node-caps'].values)
df['breast'] = breast_encoder.fit_transform(df['breast'].values)
df['breast-quad'] = breast_quad_encoder.fit_transform(df['breast-quad'].values)
df['irradiat'] = irradiat_encoder.fit_transform(df['irradiat'].values)
df['Class'] = Class_encoder.fit_transform(df['Class'].values)
df
```
